### Дополнитеьное задание


```python
corr_matrix = df[['age', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week', 'income >50K']].corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица')
plt.show()
```

Положительная корреляция между образованием и доходом: Чем выше уровень образования (education-num), тем выше вероятность заработка более 50 000 долларов в год (income >50K). Это логично, так как обычно более высокий уровень образования связан с более высокооплачиваемыми профессиями.

Положительная корреляция между часами работы и доходом: Чем больше человек работает (hours-per-week), тем выше вероятность, что его доход превышает 50 000 долларов. Это также ожидаемый результат, так как обычно более высокий доход связан с большей занятостью.

Слабая корреляция между возрастом и доходом: Хотя есть небольшая положительная корреляция, она не очень выражена. Это может говорить о том, что возраст не является единственным и определяющим фактором для высокого дохода.

Слабая корреляция остальных переменных: Между другими переменными (capital-gain, capital-loss) и доходом наблюдается слабая корреляция. Это может указывать на то, что разовые доходы или потери не оказывают существенного влияния на общий уровень дохода.
